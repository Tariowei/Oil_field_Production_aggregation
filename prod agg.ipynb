```python

import pandas as pd
import plotly.express as px
import os
os.makedirs("../images", exist_ok=True)
```

```python
data = pd.read_csv('2-production_history.txt', sep='\t')

# Convert Date column
data['Date'] = pd.to_datetime(data['Date'])

data.head()
```

```python
print(data.columns)
data.describe()
```

```python
# Convert oil from barrels to cubic meters
data['OIL_m3'] = data['OIL'] * 0.159

# Gas-Oil Ratio (GOR)
data['GOR'] = data['GAS'] / data['OIL']

# Water-Oil Ratio (WOR)
data['WOR'] = data['WATER'] / data['OIL']
```

```python
fig1 = px.area(
    data,
    x='Date',
    y='GAS',
    color='UID',
    title='Gas Production by Well'
)

fig1.show()

# Save image for GitHub
fig1.write_image("../images/gas_production.png")
```

```python
fig2 = px.scatter(
    data,
    x='Date',
    y=['GOR', 'WOR'],
    log_y=True,
    symbol='UID',
    title='GOR and WOR Trends'
)

fig2.show()

# Save image
fig2.write_image("../images/gor_wor.png")
```

```python
fig3 = px.scatter(
    data,
    x='OIL',
    y='WATER',
    color='UID',
    title='Oil vs Water Production'
)

fig3.show()

# Save image
fig3.write_image("../images/oil_vs_water.png")
![Gas Production](images/gas_production.png)
```

- Some wells exhibit declining production trends, indicating possible reservoir depletion  
- Increasing Water-Oil Ratio (WOR) suggests water breakthrough in certain wells  
- Variations in Gas-Oil Ratio (GOR) highlight differences in reservoir behaviour  
- Data-driven monitoring can support production optimisation and intervention planning  

This analysis demonstrates how production data can be used to monitor well performance and identify early signs of inefficiencies.

Such insights are valuable for production engineers in optimizing output, improving recovery, and managing reservoir performance effectively.
